```python
import pandas as pd
import numpy as np
import os
```

```python
path = os.getcwd()
for i in os.listdir(path):
    if '.csv' in i:
        print(i)
        
```

```python
main_df = pd.read_csv('DiabetesAtlasData.csv').iloc[:-1]
```

```python
Obesity_Percentage = pd.read_csv('DiabetesAtlasData_Bivariate.csv').iloc[:-1]
Obesity_Percentage = Obesity_Percentage[['County','County_FIPS','State','Obesity (Percentage) ']]
```

```python
poverty = pd.read_csv('Poverty.csv').iloc[:-1]
poverty = poverty[['County','County_FIPS','State','Children in Poverty-2021-Percent']]
```

```python
reduced_lunch = pd.read_csv('DiabetesAtlasData 8.csv').iloc[:-1]
reduced_lunch = reduced_lunch[['County','County_FIPS','State','Enrolled in Free or Reduced Lunch-2020-2021-Percent']]
```

```python
food_insecurity = pd.read_csv('DiabetesAtlasData (2).csv').iloc[:, np.r_[0:3, -1]].iloc[:-1]
healthy_foods = pd.read_csv('DiabetesAtlasData (3).csv').iloc[:, np.r_[0:3, -1]].iloc[:-1]
physican_rate = pd.read_csv('DiabetesAtlasData (4).csv').iloc[:, np.r_[0:3, -2]].iloc[:-1]
exercise_opportunities = pd.read_csv('DiabetesAtlasData (5).csv').iloc[:, np.r_[0:3, -1]].iloc[:-1]
commute = pd.read_csv('DiabetesAtlasData (6).csv').iloc[:, np.r_[0:3, -1]].iloc[:-1]
minority = pd.read_csv('DiabetesAtlasData (7).csv').iloc[:, np.r_[0:3, -1]].iloc[:-1]
free_lunch = pd.read_csv('DiabetesAtlasData 8.csv').iloc[:, np.r_[0:3, -1]].iloc[:-1]

main_df.County_FIPS = main_df.County_FIPS.astype(str).astype(float).astype(int).astype(str).str.zfill(5)
Obesity_Percentage.County_FIPS = Obesity_Percentage.County_FIPS.astype(float).astype(int).astype(str).str.zfill(5)
poverty.County_FIPS = poverty.County_FIPS.astype(str).astype(float).astype(int).astype(str).str.zfill(5)
reduced_lunch.County_FIPS = reduced_lunch.County_FIPS.astype(float).astype(int).astype(str).str.zfill(5)
food_insecurity.County_FIPS = food_insecurity.County_FIPS.astype(float).astype(int).astype(str).str.zfill(5)
healthy_foods.County_FIPS = healthy_foods.County_FIPS.astype(float).astype(int).astype(str).str.zfill(5)
physican_rate.County_FIPS = physican_rate.County_FIPS.astype(float).astype(int).astype(str).str.zfill(5)
exercise_opportunities.County_FIPS = exercise_opportunities.County_FIPS.astype(float).astype(int).astype(str).str.zfill(5)
commute.County_FIPS = commute.County_FIPS.astype(float).astype(int).astype(str).str.zfill(5)
minority.County_FIPS = minority.County_FIPS.astype(float).astype(int).astype(str).str.zfill(5)
free_lunch.County_FIPS = free_lunch.County_FIPS.astype(float).astype(int).astype(str).str.zfill(5)
```

```python
final_df = main_df.merge(Obesity_Percentage, on=['County_FIPS', 'County', 'State'])\
       .merge(poverty, on=['County_FIPS', 'County', 'State'])\
       .merge(reduced_lunch, on=['County_FIPS', 'County', 'State'])\
       .merge(food_insecurity, on=['County_FIPS', 'County', 'State'])\
       .merge(healthy_foods, on=['County_FIPS', 'County', 'State'])\
       .merge(physican_rate, on=['County_FIPS', 'County', 'State'])\
       .merge(exercise_opportunities, on=['County_FIPS', 'County', 'State'])\
       .merge(commute, on=['County_FIPS', 'County', 'State'])\
       .merge(minority, on=['County_FIPS', 'County', 'State'])\
       .merge(free_lunch, on=['County_FIPS', 'County', 'State'])

       
              
final_df
```

```python
final_df.to_csv('final_data_DVA_02212024.csv',index=False)
```
